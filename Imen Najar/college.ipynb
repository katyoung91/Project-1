## Median Debt at Graduation Charts

```python
#Read the data and convert it to a dataframe
cd = Path("Resources/clean_consolidated_data.csv")
chart_df=pd.read_csv(cd, low_memory=False)
```

### Bar plot showing Region vs. Mean Median Debt at Graduation 

```python
# Bar plot showing Region vs. Average Median Debt at Graduation 
# Data to use

Region_debt1=chart_df.set_index('Region')

Region_debt=Region_debt1.groupby('Region')[["Median Debt at Graduation"]].mean()

Region_debt=Region_debt.sort_values('Median Debt at Graduation', ascending= False)

# create the plot
Region_debt_plot =Region_debt.plot.bar()
Region_debt_plot.set_xlabel("Region")
Region_debt_plot.set_ylabel("Average Median Debt at Graduation")
Region_debt_plot.set_title ("Region vs. Average Median Debt at Graduation \n")
Region_debt_plot.legend(loc='lower left')
Region_debt_plot
plt.savefig("Resources/RegionIM.png",bbox_inches='tight')
```

```python
# ANOVA test between different groups
import scipy
df=chart_df
group0 = df[df["Region"] == 'New England']["Median Debt at Graduation"]
group1 = df[df["Region"] == 'Mid-atlantic']["Median Debt at Graduation"]
group2 = df[df["Region"] == 'Southeast']["Median Debt at Graduation"]
group3 = df[df["Region"] == 'East north central']["Median Debt at Graduation"]
group4 = df[df["Region"] == 'West North central']["Median Debt at Graduation"]
group5 = df[df["Region"] == 'North Central']["Median Debt at Graduation"]
group6 = df[df["Region"] == 'Southwest']["Median Debt at Graduation"]
group7 = df[df["Region"] == 'West']["Median Debt at Graduation"]
group8 = df[df["Region"] == 'U.S. Territories']["Median Debt at Graduation"]


results = scipy.stats.f_oneway(group0, group1, group2, group3, group4, group5,group6, group7)
results
print(results)


if results.pvalue < 0.5:
    print("We reject the null hypothesis, there is a difference")
else:
    print("We accept the null hypothesis, there is no difference")
```

```python
# ANOVA test between different groups
import scipy
df=chart_df
group0 = df[df["Region"] == 'Mid-atlantic']["Median Debt at Graduation"]
group1 = df[df["Region"] == 'East north central']["Median Debt at Graduation"]
group2 = df[df["Region"] == 'New England']["Median Debt at Graduation"]


results = scipy.stats.f_oneway(group0, group1, group2)
results
print(results)


if results.pvalue < 0.5:
    print("We reject the null hypothesis, there is a difference")
else:
    print("We accept the null hypothesis, there is no difference")
```

#### Conclusion - Region vs. Mean Median Debt at Graduation-

The Mid-Atlantic,  East North Central, New England, West North Central and Southeast have the largest average debt, at more than $20,000.  The average debt at Graduation in North Central, Southwest and West is slightly less than $20,000. Unlike the rest, the average Debt at Graduation in the  US Territories is very low, less than $10,000. 
We can conclude that  most students, regardless of their  region, have on average the same debt level–approximately $20,000–at graduation except for the US Territories.
Across all regions there are statistical differences (p-value=0.00027), but grouping specific regions together, for example the Mid-Atlantic, East North Central and New England, there is no difference in debt (p-value=0.63).

### Median Debt at Graduation vs. Debt Repayment Rate

```python
# Equation for correlation and regression

import scipy

mdg = chart_df['Median Debt at Graduation']
drr_5 = chart_df['Debt repayment rate among grads - 5 years']
#Create scatte rplot 
plt.scatter(mdg, drr_5, marker='o')
#title and labels 
plt.title( 'Median Debt at Graduation vs. Debt Repayment Rate - 5 years  \n')
plt.xlabel('Median Debt at Graduation')
plt.ylabel('Debt Repayment Rate among grads - 5 years')
#Linear Regression plots
y_value=drr_5
x_value=mdg
(slope, intercept, rvalue, pvalue,stderr)=scipy.stats.linregress (x_value, y_value)
regress_value= x_value*slope+intercept
line_eq='y='+str(round(slope,2))+'x+'+ str(round(intercept,2))
plt.annotate(line_eq,(30000,0.5),fontsize=15,color='green')
plt.plot(x_value,regress_value,'green')
r = scipy.stats.pearsonr(x_value, y_value)
print(f"The r-value is: {r[0]}")
plt.savefig("Resources/scatterIM.png")
```

#### Conclusion - Median Debt at Graduation vs. Debt Repayment Rate-

The correlation between Median Debt at Graduation and Debt Repayment Rate -- 5 years --  is very weak. The r-value is between 0.2 and 0. There is a very weak positive correlation.
The r-value is not significant, so a regression analysis is not applicable. Other factors can affect the Dept Repayment Rate but not the Median Debt at Graduation.
