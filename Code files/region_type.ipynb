```python
#dependencies
from pathlib import Path
import pandas as pd
```

```python
#Create a path to the data
path = Path("../Resources/clean_consolidated_data.csv")

#read the data and convert it to a dataframe
df=pd.read_csv(path, low_memory=False)

df.head()
```

```python
#relabeling Region data with labels

for i in df['Region'] :
    if i == 1:
        df['Region'] ="New England"
    elif i ==2:
        df['Region'] ="Mid-atlantic"
    elif i ==3:
        df['Region'] ="East north central"
    elif i ==4:
        df['Region'] ="West North central"
    elif i ==5:
        df['Region'] ="Southeast"
    elif i ==6:
        df['Region'] ="Southwest"
    elif i ==7:
        df['Region'] ="North Central"
    elif i ==8:
        df['Region'] ="West"
    elif i ==9:
        df['Region'] ="U.S. Territories"
df.head()
```

```python
#relabeling type data with labels

for i in df['Type of School'] :
    if i == 1:
        df['Type of School'] ="Public"
    elif i ==2:
        df['Type of School'] ="Private Non-profit"
    elif i ==3:
        df['Type of School'] ="Private For-Profit"
        
df.head()
```

```python
#convert to csv
df.to_csv('Resources/clean_consolidated_data.csv')
```
