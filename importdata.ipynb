```python
#This code will import the necessary data from the larger data file and consolidate it into a new data frame that will then be saved as a new file.

#dependencies
from pathlib import Path
import pandas as pd
```

```python
#Create a path to the data
original_data = Path("Resources/MERGED2011_12_PP.csv")

#read the data and convert it to a dataframe
original_df=pd.read_csv(original_data, low_memory=False)

original_df.head()
```

```python
#Create a new dataframe using just the columns of data we need (pre-determined from sifting through the file and key)

final_df =original_df[["UNITID","INSTNM","STABBR","REGION","PREDDEG","CONTROL","ADM_RATE","TUITIONFEE_IN","TUITIONFEE_OUT",
                       "DEBT_MDN","GRAD_DEBT_MDN","COMPL_RPY_1YR_RT","COMPL_RPY_3YR_RT","COMPL_RPY_5YR_RT",
                       "MN_EARN_WNE_P6","MN_EARN_WNE_P10","CIP01BACHL","CIP05BACHL","CIP09BACHL","CIP11BACHL","CIP13BACHL",
                       "CIP14BACHL","CIP23BACHL","CIP26BACHL","CIP27BACHL","CIP38BACHL","CIP42BACHL","CIP45BACHL","CIP50BACHL",
                       "CIP52BACHL","CIP54BACHL"]].copy()

final_df.head()
```

```python
#relabel column headers

final_df.columns=["UNITID","Institution Name","State","Region","Predominant Degree","Type of School","Admission Rate",
                  "Instate Tuition and Fees","Out of State Tuition and Fees","Median Debt Original Amount",
                  "Median Debt at Graduation","Debt repayment rate among grads - 1 year","Debt repayment rate among grads - 3 years",
                  "Debt repayment rate among grads - 5 years","Mean earnings 6 years after school entry",
                  "Mean earnings 10 years after school entry","Bachelor of Agriculture","Bachelor of Ethnic, Cultural or Gender Studies",
                  "Bachelor of Comms - Journalism","Bachelor of Comp Sci / IT","Bachelor of Education","Bachelor of Engineering", "Bachelor of English",
                  "Bachelor of Bio / Biomedical Sci","Bachelor of Math","Bachelor of Philosophy/Religion","Bachelor of Psychology",
                  "Bachelor of Social Science","Bachelor of Visual / Performing Arts","Bachelor of Business / Marketing","Bachelor of History"]

final_df.head()
```

```python
#drop rows with null values
final_df=final_df.dropna()

#drop universities that don't have bachelors programs for any of the fields of study we care about
indexDegree =final_df[ (final_df["Bachelor of Agriculture"] == 0) & (final_df["Bachelor of Ethnic, Cultural or Gender Studies"] == 0) 
                      & (final_df["Bachelor of Comms - Journalism"] == 0) & (final_df["Bachelor of Comp Sci / IT"] == 0)
                      & (final_df["Bachelor of Education"] == 0) & (final_df["Bachelor of Engineering"] == 0)
                      & (final_df["Bachelor of English"] == 0) & (final_df["Bachelor of Bio / Biomedical Sci"] == 0)
                      & (final_df["Bachelor of Math"] == 0) & (final_df["Bachelor of Philosophy/Religion"] == 0)
                      & (final_df["Bachelor of Psychology"] == 0) & (final_df["Bachelor of Social Science"] == 0)
                      & (final_df["Bachelor of Visual / Performing Arts"] == 0) & (final_df["Bachelor of Business / Marketing"] == 0)
                      & (final_df["Bachelor of History"] == 0)
                      ].index
final_df.drop(indexDegree, inplace=True)
```

```python
#export file
final_df.to_csv('Resources/consolidated_data.csv')
```
